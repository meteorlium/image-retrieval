查看图片和 caption

```python
import json
import os

base_path = "../"
```

```python
json_data = json.load(open(os.path.join(base_path, "datasets", "ShareGPT4V", "sharegpt4v_instruct_gpt4-vision_cap100k.json")))

print(json_data[0])
```

```python
idx = 100

image_info = json_data[idx]

image_info
```

```python
# show caption

caption = image_info["conversations"][1]["value"]

print(caption)
```

```python
# print image
image_path = image_info["image"]
if image_path.startswith("coco/train2017"):
    image_path = image_path.replace("coco/train2017", "datasets/coco/images/train2017")
    image_path = os.path.join(base_path, image_path)
    if not os.path.exists(image_path):
        image_path = "../" + image_path
else:
    raise NotImplementedError(f"image_path: {image_path}")

image_path
```

```python
# show image - 使用PIL Image直接显示，逻辑更简单

from PIL import Image

# 直接读取并显示图片
image = Image.open(image_path)
print(f"图片尺寸: {image.size}")

# 在notebook中直接显示
display(image)
```
